```python
import requests

url = "https://www.google.com/maps/place/%EC%84%9C%EC%9A%B8%EC%97%AD+%EA%B3%B5%ED%95%AD%EC%B2%A0%EB%8F%84/data=!3m1!4b1!4m6!3m5!1s0x357ca315e6d51773:0x2ea3b0bd6322ddd5!8m2!3d37.5535242!4d126.9696972!16s%2Fg%2F11s4339c88?entry=ttu"

rm = requests.get(url)
print(type(rm))
print(rm.cookies)
print(rm.headers)
print(rm.encoding)
print(rm.status_code)
print(rm.json)
#print(rm.text)
```

```python
#print(rm.text)
#print(rm.json)
print(rm.content)

from bs4 import BeautifulSoup

soup = BeautifulSoup(rm.content)
print(soup.prettify())
```
