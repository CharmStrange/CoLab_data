## 01. 데이터셋 불러오기

```python
import pandas as pd

df = pd.read_csv("/content/FastFoodNutritionMenuV2.csv")
df
```

## 02. 데이터셋 파악하기

```python
# 열 정보 추출
for column in df.columns:
    print(column)
```

- **Item** : **상품 명**

- **Calories from Fat** : **열량 중 지방**
- **Total Fat** : **총 지방**(g)
- **Saturated Fat** : **포화지방**(g)
- **Trans Fat** : **트랜스지방**(g)
- **Cholesterol** : **콜레스테롤**(mg)
- **Sodium** : **나트륨**(mg)
- **Carbs** : **탄수화물**(g)
- **Fiber** : **식이섬유**(g)
- **Sugars** : **당**(g)
- **Protein** : **단백질**(g)
- **Weight Watchers Pnts** : **영양 성분 평가 지표**

## 01. 결측치 확인하고 처리하기

```python
print(df.isna().sum())

print("\n결측치 비율")
for column in df.isna().sum():
    print(f"{column / len(df) * 100}%")
```
