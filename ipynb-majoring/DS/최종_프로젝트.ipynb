# Data Wrangling

## 필요한 정보만 골라내고, 데이터 정리하기

## 01. 결측치 확인하고 처리하기

```python
# 기본 정보 체크
print(df.isna().sum())

print("\n결측치 비율")
for column in df.isna().sum():
    print(f"{column / len(df) * 100}%")
```

분석에 필요하지 않은 정보는 버린다.

`Calories from Fat`, `Weight Watchers Pnts`

```python
# 다른 결측치를 확인
df = df.drop(['Calories from\nFat', 'Weight Watchers\nPnts'], axis=1)
missing_values = df[df.isnull().any(axis=1)]
missing_values
```

필요하지 않은 열을 제외한 나머지 열의 결측치를 보니 전부 탄산음료, 차 등의 음료고, 이런 음료는 보통 단백질, 탄수화물 등을 포함하지 않으니 0으로 처리를 해 주었다.

```python
# 나머지 결측치 처리
new_df = df.fillna(0)
new_df.isna().sum()
```

## 02. 데이터 정리하기

```python
# 필요한 열에 대한 정리
new_df[['Company', 'Item', 'Saturated Fat\n(g)', 'Cholesterol\n(mg)', 'Sodium \n(mg)', 'Sugars\n(g)']]

# 데이터 형태 확인
print(new_df['Saturated Fat\n(g)'].dtype)
print(new_df['Cholesterol\n(mg)'].dtype)
print(new_df['Sodium \n(mg)'].dtype)
print(new_df['Sugars\n(g)'].dtype)

# 데이터 형태 변환
new_df['Saturated Fat\n(g)'] = pd.to_numeric(new_df['Saturated Fat\n(g)'].str.replace(r'[^0-9.]', ''), errors='coerce').fillna(0).astype(float)
new_df['Cholesterol\n(mg)'] = pd.to_numeric(new_df['Cholesterol\n(mg)'].replace('\xa0', ''), errors='coerce').fillna(0).astype(float)
new_df['Sodium \n(mg)'] = pd.to_numeric(new_df['Sodium \n(mg)'].replace('\xa0', ''), errors='coerce').fillna(0).astype(float)
new_df['Sugars\n(g)'] = pd.to_numeric(new_df['Sugars\n(g)'].replace('\xa0', ''), errors='coerce').fillna(0).astype(float)

# 데이터 형태 확인
print(new_df['Saturated Fat\n(g)'].dtype)
print(new_df['Cholesterol\n(mg)'].dtype)
print(new_df['Sodium \n(mg)'].dtype)
print(new_df['Sugars\n(g)'].dtype)
```

#Data Analysis

## 탐색적 데이터 분석, 유의미한 결과 도출

## 01.전체적인 상관 관계와 데이터 분포 확인

```python
# 각 열에 대해, 패스트 푸드 기업 별로 묶어 정보를 출력
print(new_df[['Company', 'Saturated Fat\n(g)']].groupby('Company').describe())
print(new_df[['Company', 'Cholesterol\n(mg)']].groupby('Company').describe())
print(new_df[['Company', 'Sodium \n(mg)']].groupby('Company').describe())
print(new_df[['Company', 'Sugars\n(g)']].groupby('Company').describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
# 상관 관계 시각화
plt.figure(figsize=(10, 8))
mask = np.triu(new_df.corr())
sns.heatmap(new_df.corr(), annot=True, cmap='coolwarm', linewidths=.5, mask=mask)
plt.title('Correlation Plot')
plt.show()
```

```python
# 상관 관계와 유의 확률 확인
for column1 in new_df[['Saturated Fat\n(g)', 'Cholesterol\n(mg)', 'Sodium \n(mg)', 'Sugars\n(g)']].columns:
    for column2 in new_df[['Saturated Fat\n(g)', 'Cholesterol\n(mg)', 'Sodium \n(mg)', 'Sugars\n(g)']].columns:
        if column1 != column2:
            correlation, p_value = stats.pearsonr(new_df[['Saturated Fat\n(g)', 'Cholesterol\n(mg)', 'Sodium \n(mg)', 'Sugars\n(g)']][column1], new_df[['Saturated Fat\n(g)', 'Cholesterol\n(mg)', 'Sodium \n(mg)', 'Sugars\n(g)']][column2])
            print(f"Correlation between {column1} and {column2}: {correlation:.4f}, p-value: {p_value:.4f}\n")
```

상관 관계에 대한 유의 확률이 모두 0.05보다 낮으니, 서로 유의미한 관계가 있다고 볼 수 있다.

```python
# 단위 별로 묶어 상자 도표 확인
sns.boxplot(data=new_df[['Saturated Fat\n(g)', 'Sugars\n(g)']])
plt.title('Saturated Fat & Sugars (g)')
plt.show()

sns.boxplot(data=new_df[['Cholesterol\n(mg)', 'Sodium \n(mg)']])
plt.title('Cholesterol & Sodium (mg)')
plt.show()
```

## 02. 분석 기준을 정하고 그에 맞게 데이터를 확인

분석 기준은, 각 패스트 푸드의 영양 성분별로 기준치를 정한 것이고 총 네 개의 그래프를 그려 확인한다.

```python
# 조건 설정
"""
당 : 35g 이하
나트륨 : 2300mg 이하
포화 지방 : 15g 이하
콜레스테롤 : 300mg 이하
"""
condition = (new_df['Saturated Fat\n(g)'] > 15) & (new_df['Cholesterol\n(mg)'] > 300) & \
            (new_df['Sodium \n(mg)'] > 2300) & (new_df['Sugars\n(g)'] > 35)

plt.bar(new_df.index, new_df['Saturated Fat\n(g)'], color='blue', label='Saturated Fat (g)')
plt.bar(new_df[condition].index, new_df[condition]['Saturated Fat\n(g)'], color='red')
plt.axhline(y=15, color='red', linestyle='--', label='RDI')
plt.xlabel('All Fast Foods')
plt.ylabel('Saturated Fat (g)')
plt.title('Saturated Fat Bar Plot')
plt.legend()
plt.show()

plt.bar(new_df.index, new_df['Cholesterol\n(mg)'], color='blue', label='Cholesterol (mg)')
plt.bar(new_df[condition].index, new_df[condition]['Cholesterol\n(mg)'], color='red')
plt.axhline(y=300, color='red', linestyle='--', label='RDI')
plt.xlabel('All Fast Foods')
plt.ylabel('Cholesterol (mg)')
plt.title('Cholesterol Bar Plot')
plt.legend()
plt.show()

plt.bar(new_df.index, new_df['Sodium \n(mg)'], color='blue', label='Sodium (mg)')
plt.bar(new_df[condition].index, new_df[condition]['Sodium \n(mg)'], color='red')
plt.axhline(y=2300, color='red', linestyle='--', label='RDI')
plt.xlabel('All Fast Foods')
plt.ylabel('Sodium (mg)')
plt.title('Sodium Bar Plot')
plt.legend()
plt.show()

plt.bar(new_df.index, new_df['Sugars\n(g)'], color='blue', label='Sugars (g)')
plt.bar(new_df[condition].index, new_df[condition]['Sugars\n(g)'], color='red')
plt.axhline(y=35, color='red', linestyle='--', label='RDI')
plt.xlabel('All Fast Foods')
plt.ylabel('Sugars (g)')
plt.title('Sugars Bar Plot')
plt.legend()
plt.show()
```
